```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
filename = '../DATA/BYAN.csv'
df = pd.read_csv(filename, sep=';', index_col='date')
```

```python
df['SMA50'] = df['close'].rolling(40).mean()
df['SMA100'] = df['close'].rolling(80).mean()
df['SMA200'] = df['close'].rolling(160).mean()
df.dropna(inplace=True)
```

```python
df[['close', 'SMA50', 'SMA100', 'SMA200']].plot(label='RELIANCE', figsize=(16, 8))
```
