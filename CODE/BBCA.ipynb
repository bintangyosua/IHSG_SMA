```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
filename = '../DATA/BBCA.csv'
df = pd.read_csv(filename, sep=';')
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')

df['tahun'] = df['date'].dt.year
df['bulan'] = df['date'].dt.month

results = df.groupby(['tahun', 'bulan']).mean()
```

```python
results['SMA30'] = results['close'].rolling(2).mean()
results.dropna(inplace=True)
```

```python
results[['close', 'SMA30']].plot(label='RELIANCE', figsize=(16, 8))
```
